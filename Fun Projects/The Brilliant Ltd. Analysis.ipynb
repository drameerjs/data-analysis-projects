# Data preprocessing
Loading the necessary libraries and datasets:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import json
```

```python
tbl_df = pd.read_csv('C:/Users/ameer/Desktop/Practicum/The Brilliant Ltd/sim-companies-account-history_1.csv')
tbl_df.tail(50)
```

Looking for duplicates or missing values:

```python
print('Duplicated rows:', tbl_df.duplicated().sum())
print('Missing values:\n', tbl_df.isna().sum())
```

No duplicates or missing values as expected - lets move on!
**Next**: we'll make some changes in our dataframe regarding column names and adding few additional information/columns

```python
tbl_df['Timestamp'] = pd.to_datetime(tbl_df['Timestamp']).dt.strftime('%Y-%m-%d %H:%M:%S')
tbl_df['Date'] = tbl_df['Timestamp'].astype('datetime64[D]')
tbl_df['Month'] = tbl_df['Date'].astype('datetime64[M]').dt.strftime('%Y-%m')
tbl_df
```

```python
tbl_df['Description'].value_counts()
```

Above we've added the 'Date' and 'Month' columns so we can be able to perform analysis based on those features.

**Next:** we'll look into the details and description

```python
def update_desc(desc):
    if ('Production of' in desc) or ('Sales of' in desc) or ('Achievement' in desc):
        desc = desc.split()
        return ' '.join(desc[2:])
    elif ('market order partially filled' in desc) or ('market order filled' in desc):
        desc = desc.split()
        return desc[0]
    elif 'Bought' in desc:
        desc = desc.split()
        return desc[1]
    elif 'contract' in desc:
        desc = desc.replace('contract signed by', '-')
        return desc
    else:
        return desc
        

tbl_df['Description'] = tbl_df['Description'].apply(update_desc)
tbl_df
```

We've successfully made the descriptions much more compact, with the help of the categories, we'll know if each row is a market, sale, fees or production!

**Next:** we want to add a profit column, to calculate later the total profit made by selling items.

```python
def show_profit(details):
    details = json.loads(details)
    if details.get('profit'):
        return float(str(details['profit']).strip('$'))
    else:
        return 0
    
def show_amount(details):
    details = json.loads(details)
    if details.get('amount'):
        return int(details['amount'])
    elif details.get('quantity'):
        return int(details['quantity'])
    else:
        return 0

def show_unit_price(details):
    details = json.loads(details)
    if details.get('price'):
        return float(str(details['price']).strip('$'))
    else:
        return 0

tbl_df['Profit'] = tbl_df['Details'].apply(show_profit)
tbl_df['Amount'] = tbl_df['Details'].apply(show_amount)
tbl_df['Unit Price'] = tbl_df['Details'].apply(show_unit_price)
tbl_df.head()
```

Lastly, for easier readability, we are going to add a function to add commas after every 3 digits in each number (a thousands' comma) - we'll keep this function for later use:

```python
def add_commas(column):
    column['Money'] = '${:,.2f}'.format(column['Money'])
    column['Profit'] = '${:,.2f}'.format(column['Profit'])
    column['Amount'] = '{:,.0f}'.format(column['Amount'])
    return column
```

Now that we have done enough changes to our data, we can move on and start our data analysis.

### Rocket Fuel Analysis - P&E
Rocket fuel takes both methane and power for production. Therefore, the numbers that we'll be seeing will include money already spent on Power and Methane (other than worker wages):

Our company rocket fuel production from 29th of Aug till the 10th of Sep was actually 0 but there's no records for dates with 0 production therefore we can see the jump in straight line from 29th of Aug to the 10th of Sep without any special changes in-between.
